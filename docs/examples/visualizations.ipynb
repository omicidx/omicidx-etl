# Data Visualizations

This notebook demonstrates how to create visualizations from OmicIDX data.


```python
import duckdb
import pandas as pd
import plotly.express as px

conn = duckdb.connect('omicidx.duckdb')
```

## Study Counts Over Time


```python
# Get study counts by month
df = conn.execute("""
    SELECT 
        DATE_TRUNC('month', publish_date) as month,
        COUNT(*) as study_count
    FROM bronze.stg_sra_studies
    WHERE publish_date >= '2020-01-01'
    GROUP BY month
    ORDER BY month
""").df()

# Create visualization
fig = px.line(df, x='month', y='study_count', 
              title='SRA Studies Published Over Time')
fig.show()
```
