# Common Query Examples

This notebook demonstrates common query patterns for the OmicIDX data warehouse.


```python
import duckdb
import pandas as pd

# Connect to the database
# Note: Download omicidx.duckdb first from the deployment URL
conn = duckdb.connect('omicidx.duckdb')
```

## List Available Tables


```python
# List schemas
schemas = conn.execute("SHOW SCHEMAS").df()
print("Available schemas:")
print(schemas)

# List tables in bronze schema
tables = conn.execute("SHOW TABLES FROM bronze").df()
print("\nTables in bronze schema:")
print(tables)
```

## SRA Queries


```python
# Count studies by type
result = conn.execute("""
    SELECT 
        study_type,
        COUNT(*) as count
    FROM bronze.stg_sra_studies
    GROUP BY study_type
    ORDER BY count DESC
    LIMIT 10
""").df()

print(result.to_markdown())
```
